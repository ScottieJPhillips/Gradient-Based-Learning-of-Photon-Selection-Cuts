```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import torch

def data_mask(pt_ranges, mu_ranges, d):
    mask = (d['y_pt_custom'] >= pt_ranges[0]) & (d['y_pt_custom'] < pt_ranges[1]) & \
           (d['evt_mu'] >= mu_ranges[0]) & (d['evt_mu'] < mu_ranges[1]) & (abs(d['y_eta_custom']) < 0.8 )
    return d[mask]

def x_y_splits(pt_ranges, mu_ranges, sgn, bkg, features):
    x_train = [[None for _ in range(len(mu_ranges))] for _ in range(len(pt_ranges))]
    x_test  = [[None for _ in range(len(mu_ranges))] for _ in range(len(pt_ranges))]
    y_train = [[None for _ in range(len(mu_ranges))] for _ in range(len(pt_ranges))]
    y_test  = [[None for _ in range(len(mu_ranges))] for _ in range(len(pt_ranges))]
    sc_list = [[StandardScaler() for _ in range(len(mu_ranges))] for _ in range(len(pt_ranges))]

    for i, pt_range in enumerate(pt_ranges):         
        for j, mu_range in enumerate(mu_ranges):    
            sgn_data = data_mask(pt_range, mu_range, sgn)
            bkg_data = data_mask(pt_range, mu_range, bkg)
            
            max_events_per_class = 25000
            if len(sgn_data) > max_events_per_class:
                sgn_data = sgn_data[:max_events_per_class]
            if len(bkg_data) > max_events_per_class:
                bkg_data = bkg_data[:max_events_per_class]

            
            print(f'for pt = {pt_range} and mu = {mu_range}')
            print(f'SGN:{len(sgn_data)}, BKG:{len(bkg_data)}')
            if len(sgn_data) ==0 and len(bkg_data) == 0:
                continue
            X_sgn = np.column_stack([sgn_data[f].to_numpy() for f in features]) if len(sgn_data) > 0 else np.empty((0, len(features)))
            X_bkg = np.column_stack([bkg_data[f].to_numpy() for f in features]) if len(bkg_data) > 0 else np.empty((0, len(features)))
            y_sgn = np.ones(len(X_sgn))
            y_bkg = np.zeros(len(X_bkg))

            X = np.concatenate([X_sgn, X_bkg], axis=0)
            y = np.concatenate([y_sgn, y_bkg], axis=0)

            stratify_y = y if len(np.unique(y)) > 1 else None

            X_train_split, X_test_split, Y_train_split, Y_test_split = train_test_split(
                X, y, test_size=0.3, random_state=42, stratify=stratify_y
            )

            scaler = sc_list[i][j]
            X_train_scaled = scaler.fit_transform(X_train_split)
            X_test_scaled = scaler.transform(X_test_split)

            x_train[i][j] = torch.tensor(X_train_scaled, dtype=torch.float32).detach()
            x_test[i][j]  = torch.tensor(X_test_scaled, dtype=torch.float32).detach()
            y_train[i][j] = torch.tensor(Y_train_split, dtype=torch.float32).detach()
            y_test[i][j]  = torch.tensor(Y_test_split, dtype=torch.float32).detach()
            # print(len(x_test[i][j]),len(y_test[i][j]))
    return x_train, x_test, y_train, y_test, sc_list


features = ['y_Rhad','y_Reta','y_Rphi','y_weta2','y_Eratio','y_deltae','y_wtots1','y_fracs1','y_weta1']

x_train, x_test, y_train, y_test, scalers = x_y_splits(pt_ranges, mu_ranges, sgn, bkg, features)

```

```python
from sklearn.metrics import auc

def compute_roc_auc(bkg_effics, sig_effics, pt_ranges, mu_ranges):
    auc_values = [[[] for _ in range(len(mu_ranges))] for _ in range(len(pt_ranges))]
    for m in range(len(pt_ranges)):
        for n in range(len(mu_ranges)):
            fpr = [0] + bkg_effics[m][n] + [1]
            tpr = [0] + sig_effics[m][n] + [1]
            roc_auc = auc(fpr, tpr)
            auc_values[m][n].append(roc_auc)
    return auc_values

auc_results = compute_roc_auc(bkg_effics, sig_effics, pt_ranges, mu_ranges)
```
